```python
# source: https://www.geeksforgeeks.org/opencv-motion-blur-in-python/
# loading library
from google.colab.patches import cv2_imshow
import cv2
import numpy as np

img = cv2.imread ('/crack_dataset/print_test1.jpg')

# Specify the kernel size
kernel_size = 15

# Create the vertical kernel
kernel_v = np.zeros((kernel_size, kernel_size))

# Fill the middle row with ones
kernel_v[:, int((kernel_size - 1)/2)] = np.ones(kernel_size)

# Normalize
kernel_v /= kernel_size

# Apply the vertical kernel
vertical_mb = cv2.filter2D(img, -1, kernel_v)

# Save the outputs
cv2.imwrite('test_vertical.jpg', vertical_mb)

# Print out
img2 = cv2.imread('test_vertical.jpg', cv2.IMREAD_UNCHANGED)
cv2_imshow(img2)

```
